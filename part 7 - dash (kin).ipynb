```python
import dash
import dash_html_components as html
import dash_core_components as dcc
import plotly.graph_objects as go
import plotly.express as px
```

```python
app = dash.Dash()   #initialising dash app
df = px.data.stocks() #reading stock price dataset

def stock_prices():
    # Function for creating line chart showing Google stock prices over time
    fig = go.Figure([go.Scatter(x = df['date'], y = df['GOOG'],\
                     line = dict(color = 'firebrick', width = 4), name = 'Google')
                     ])
    fig.update_layout(title = 'Prices over time',
                      xaxis_title = 'Dates',
                      yaxis_title = 'Prices'
                      )
    return fig


app.layout = html.Div(id = 'parent', children = [
    html.H1(id = 'H1', children = 'Styling using html components', style = {'textAlign':'center',\
                                            'marginTop':40,'marginBottom':40}),


        dcc.Graph(id = 'line_plot', figure = stock_prices())
    ]
                     )
```

```python
if __name__ == '__main__':
    app.run_server()
```

```python
import dash
import dash_core_components as dcc
import dash_html_components as html
from dash.dependencies import Input, Output
import plotly.graph_objs as go
import datetime

app = dash.Dash(__name__)

app.layout = html.Div([
    html.H1("Age vs. Time Line Graph"),
    html.Label("Enter your age:"),
    dcc.Input(id='user-age', type='number', value=25),
    dcc.Graph(id='age-graph'),
])

@app.callback(
    Output('age-graph', 'figure'),
    [Input('user-age', 'value')]
)
def update_graph(user_age):
    x = [datetime.datetime.now() - datetime.timedelta(days=i) for i in range(user_age)]
    y = list(range(user_age))

    trace = go.Scatter(x=x, y=y, mode='lines', marker={'color': 'blue'})

    layout = go.Layout(
        title=f"Age: {user_age} vs. Time",
        xaxis=dict(title='Time'),
        yaxis=dict(title='Age')
    )

    return {'data': [trace], 'layout': layout}

if __name__ == '__main__':
    app.run_server(debug=True)
```
